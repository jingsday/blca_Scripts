```python
cleaned_strings= list(blca.columns)
```

```python
intersection = list(set(list(STVs.index)).intersection(cleaned_strings))
```

```python
STVs=STVs.loc[intersection]
```

```python
STVs
```

```python
blca_lm=blca[intersection]
```

```python
display(blca_lm)
```

```python
states = STVs.columns.to_list()
print("States:", states)
DPD_df = pd.DataFrame(
    np.zeros((len(blca_lm), len(states))),
    index = list(blca_lm.index),
    columns = states,
)
```

```python
for exp_id in DPD_df.index:
    for state in states:
        DPD_df.loc[exp_id, state] = np.dot(blca_lm.loc[exp_id], STVs.loc[:, state])

display(DPD_df)
```

```python
cell_line_ids=['ACH-000142','ACH-000593','ACH-000396','ACH-000242','ACH-000384',
               'ACH-000018','ACH-000724']

DPD_cell_lines= DPD_df.loc[cell_line_ids]
DPD_cell_lines['cell_line_name']=['CAL29','BC3C','J82','RT4','SW780','T24','HT1376']
```
