```python
info_dir= '/Users/lidiayung/PhD_project/project_UCD_blca/blca_DATA/blca_DATA_LINCS/'
modules_df = pd.read_excel(
    os.path.join(info_dir,'ALL_DATA_2020_Jing.xlsx'), sheet_name = "modules", index_col = 0)
display(modules_df)

IC50_df = pd.read_excel(
    os.path.join(info_dir, "ALL_DATA_2020_Jing.xlsx"), sheet_name = "IC50s")
IC50_df.drop(columns=['Unnamed: 3','Unnamed: 4'],inplace=True)
display(IC50_df)
```

```python
results_bc3c_perturbed_df['Drug'] = 0
for i in results_bc3c_perturbed_df.index:
    # Directly update the DataFrame
    results_bc3c_perturbed_df.loc[i, 'Drug'] = sig_info.loc[i, 'pert_drug']

display(results_bc3c_perturbed_df)
```

```python
#results_bc3c_perturbed_df_reset = results_bc3c_perturbed_df.reset_index()

# Merge DataFrames on 'Drug'
results_bc3c_perturbed_df = results_bc3c_perturbed_df.reset_index().merge(IC50_df[['Drug', 'Module']], on='Drug', how='left')

# Set the index back to the original index if needed
results_bc3c_perturbed_df.set_index('index', inplace=True)

# Optionally, rename the index column if needed
results_bc3c_perturbed_df.index.name = None

# Display the resulting DataFrame
display(results_bc3c_perturbed_df)
```
